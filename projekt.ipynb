#####  3.1.2. INFORMACIJSKA DOBIT

```python
# Function to calculate entropy
def entropy(class_probs):
    return -np.sum(class_probs * np.log2(class_probs))

# Function to calculate information gain
def information_gain(data, feature_name, target_name):
    total_entropy = entropy(data[target_name].value_counts(normalize=True))

    # Calculate weighted average entropy for the feature
    weighted_feature_entropy = 0
    for value in data[feature_name].unique():
        subset = data[data[feature_name] == value]
        value_prob = len(subset) / len(data)
        value_entropy = entropy(subset[target_name].value_counts(normalize=True))
        weighted_feature_entropy += value_prob * value_entropy

    # Calculate information gain
    info_gain = total_entropy - weighted_feature_entropy
    return info_gain
```

```python
target_column = 'Custom Group' 
information_gains = {}

# Calculate information gain for each feature
for column in X_rep.drop(['ASF', 'CDR'], axis=1, inplace=False).columns:
    if column != target_column:
        information_gains[column] = information_gain(X_rep, column, target_column)

# Print information gains for each feature
print("Information Gains (predicting Custom Group):")
for feature, gain in information_gains.items():
    print(f"{feature:>20}: {gain:.4f}")
```

Prikazan je ispis informacijske dobit s obzirom na značajku Custom Group.

Značajke s većom informacijskom dobiti imaju veći utjecaj na klasifikaciju primjera u određenu klasu. U ovom skupu podataka najveću informacijsku dobit imaju eTIV i nWBV što sa medicinskog aspekta ima smisla.

Metrike koje su korištene u radu su:
- točnost (eng. accuracy) - omjer točno klasificiranih primjera u odnosu na sve primjere
$$
  accuracy  = \frac{TP + TN}{TP + FP + TN + FN}
$$
- preciznost (eng. precision) - izvedba pojedine klase, tj. omjer točno predviđenih pozitivnih primjera i ukupno pozitivnih primjera.
$$
  precision  = \frac{TP}{TP + FP}
$$

- odziv (eng. recall) - omjer uzoraka klase koja je točno predviđena.
$$
  recall = \frac{TP}{TP + FN}
$$
- F1 vrijednost -  harmonijsku sredinu preciznosti i recalla.
$$
  F1 \ score  = \frac{2*Precision*Recall}{Precision + Recall}
$$

Za svaki model prikazana je i matrica zabune.

Treniranje se provodi tako da se skup podataka podijeli u omjeru 80%:20% na skup za učenje i skup za testiranje.

Metrike se izračunavaju pomoću treniranog modela na skupu za testiranje.

#### 3.3. STABLA ODLUKE (DECISON TREE)

#### 3.6. VOTING CLASSIFIER

Voting klasifikator nije klasifikator sam po sebi, već služi kao omotač oko više različitih klasifikatora. Njegova predikcija temelji se na većinskoj klasifikaciji primjera u pojedinom klasifikatoru unutar "omotača". U hard voting varijanti primjer se svrstava u klasu koja ima najviše glasova, dok se u soft voting varijanti zbrajaju vektori vjerojatnosti pripadnosti primjera pojedinoj klasi za svaki klasifikator te se primjer klasificira kao primjer klase koja ima najveću vjerojatnost. Optimizacija hiperparametra voting klasifikatora svodi se na optimizaciju hiperparametra svakog pojeidnog klasifikatora u omotaču.  

U ovoj primjeri kao klasifikatore ćemo koristiti: 
- logističku regresiju (hiperparametri: regularaziacijski faktor)
- KNN (hiperparametar: broj susjeda)
- naivni Bayesov klasifikator

##### TRENIRANJE MODELA - HARD VOTING

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import VotingClassifier

# Define classifiers to be wrapped
gnb_clf = GaussianNB()
lr_clf = LogisticRegression(max_iter=1000)
knn_clf = KNeighborsClassifier()

hv_parameters_to_optimize = {
   'gnb__var_smoothing': [1e-9, 1e-8, 1e-7],
   'knn__n_neighbors': range(1,30, 2),
   'lr__C': [0.1, 0.5, 1, 10]
}

hv_classifier = VotingClassifier(estimators = [
    ('gnb', gnb_clf),
    ('lr', lr_clf),
    ('knn', knn_clf),
], voting='hard')


# Optimize the model
grid_search_hv = GridSearchCV(hv_classifier, hv_parameters_to_optimize, n_jobs=8, cv=k, scoring='f1')
grid_search_hv.fit(X_train, y_train)

# Best parameters
hv_optimized_smoothing = grid_search_hv.best_params_['gnb__var_smoothing']
hv_optimized_k = grid_search_hv.best_params_['knn__n_neighbors']
hv_optimized_C = grid_search_hv.best_params_['lr__C']
print("Optimized parameters (Hard voting): ")
print(f'{"var_smoothing":>20}:{hv_optimized_smoothing:>8}')
print(f'{"n_neighbors (KNN)":>20}:{hv_optimized_k:>9}')
print(f'{"C (LR)":>20}: {str(hv_optimized_C):>8}')
```

```python
# Optimized Voting classifier
hv_optimized_classifier = VotingClassifier(estimators = [
    ('gnb', GaussianNB(var_smoothing=hv_optimized_smoothing)),
    ('lr', LogisticRegression(max_iter=1000, C=hv_optimized_C)),
    ('knn', KNeighborsClassifier(n_neighbors=hv_optimized_k)),
], voting='hard')


hv_optimized_classifier.fit(X_train, y_train)

# Prediction on test set
y_pred_hv = hv_optimized_classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred_hv)
print("Accuracy (optimized):", accuracy)
```

##### VALIDACIJA MODELA - HARD VOTING

```python
y_pred = hv_optimized_classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Add to metrics_table
metric_table.loc[len(metric_table)] = ['Voting (hard)', accuracy, precision, recall, f1]

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
```

Dobivene razlike u metrikama moguće je objasniti dvjema činjenicama. U prvom redu usporedbom dobivenih matrica zabune i matrica zabune iz polaznog rada jasno je da ispitni skup nije jednak. Razlog ove razlike je neprecizno dokumentrianje metode podjele na skup za ispitivanje i skup za učenje u polaznom radu. Drugi faktor koji utječe na razlike u dobivnem metrikama jest optimizacija hiperparametra. Iz istraživačkog rada nije jasno koji hiperparametri modeli su bili optimirani pa je moguće da se hiperparametri koji su optimirani u ovom radu razlikuju od hiperparametara koji su optimirani u polaznom radu.
